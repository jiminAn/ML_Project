```python
LABELED_DISAST_DATA_PATH = '../data/disaster_labeled.csv'
LABELED_NONDISAST_DATA_PATH = '../data/non-disaster_labeled.csv'
ORIGIN_TRAIN_DATA_PATH = '../data/train.csv'
ORIGIN_TEST_DATA_PATH = '../data/test.csv'
```

```python
import pandas as pd
```

```python
df = pd.read_csv(LABELED_DISAST_DATA_PATH, encoding='euc-kr')
df.head()
```

```python
# df.drop(labels='Unnamed: 4', axis=1, inplace=True)
df.drop(labels=['Unnamed: 5', 'Unnamed: 4', 'idx'], axis=1, inplace=True)
df.rename(columns={"label": "target"}, inplace=True)
```

```python
df
```

```python
df.to_csv("./data/labeled_disaster_data.csv")
```

```python
df = pd.read_csv(LABELED_NONDISAST_DATA_PATH, encoding='euc-kr')
df
```

```python
df.drop(labels='index', axis=1, inplace=True)
df.rename(columns={"label": "target"}, inplace=True)
df
```

```python
df.to_csv('./data/labeled_nondisaster_data.csv')
```

```python
df = pd.read_csv(ORIGIN_TRAIN_DATA_PATH)
df
```

```python
df.drop(labels='id', axis=1, inplace=True)
df.drop(labels='location', axis=1, inplace=True)
df
```

```python
df.to_csv('./data/origin_train.csv')
```

```python
df = pd.read_csv(ORIGIN_TEST_DATA_PATH, encoding='utf-8')
df
```
